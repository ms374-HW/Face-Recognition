```python
import PIL.Image
import PIL.ImageDraw
from IPython.display import display # to display images
import face_recognition
```

```python
image = face_recognition.load_image_file("../input/pins-face-recognition/105_classes_pins_dataset/pins_Adriana Lima/Adriana Lima0_0.jpg")
```

```python
# returns list of faces found with 4 coordinates
face_locations = face_recognition.face_locations(image)
```

```python
nfaces = len(face_locations)
print(f"There are {nfaces} face(s) in this photograph.")
```

```python
pil_image = PIL.Image.fromarray(image)
```

```python
for face_location in face_locations:
    top, right, bottom, left = face_location
    print(f"pixel location Top: {top}, Left: {left}, Bottom: {bottom}, Right: {right}")
    
    # Box around face
    draw = PIL.ImageDraw.Draw(pil_image)
    draw.rectangle([left, top, right, bottom], outline="red")

```

```python
# Display the image on screen
display(pil_image)
```
